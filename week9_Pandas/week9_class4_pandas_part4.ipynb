Let's do another task , let's take the titanic data , now group by the data based on who has survived and who has not survived

```python
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df_titanic=pd.read_csv(url)
```

```python
df_titanic
```

```python
#now let's groupby based on who survived
g=df_titanic.groupby('Survived')
```

```python
#now see who survived and who didn't survived their fare sum:
#g is object
g.sum()
```

```python
g.mean()
```

```python
#now see who paid maximum in a group:
g1.max()
```

```python
#Now see it in rotated form
g1.max().T
#row column interchange , transpose
```

Jekhane fare:   
10 euro er niche lekho low

10-20 euro mid

20 er upor high

```python
def cat_fare(x):
  if x<10:
    return 'cheap'
  if x>=10 and x<20:
    return 'mid'
  else:
    return 'high'
```

```python
df_titanic['category_Fare']=df_titanic['Fare'].apply(cat_fare)
```
