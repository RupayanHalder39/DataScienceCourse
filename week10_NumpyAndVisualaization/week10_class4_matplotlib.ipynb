Histogram plot:

Imagine you have a big jar filled with colorful candies. You want to know how many candies you have of each color. To find out, you decide to make a histogram.

A histogram is like a special chart that shows you how many things fall into different categories. In our case, the categories are the colors of the candies. You take out one candy at a time and count how many candies you have of each color.

Let's say you have red, blue, green, and yellow candies. You start counting and notice that you have 5 red candies, 3 blue candies, 7 green candies, and 2 yellow candies. You make a little bar chart where each color has a bar and the height of the bar represents the number of candies you have.

The histogram helps you see at a glance which color has more candies and which color has fewer candies. You might notice that you have the most green candies, while the yellow candies are the least.

Histograms are not just for candies. They can be used to show how many times something happens in different categories. For example, you can use a histogram to show how many students got different grades on a test or how many cars pass by on different days of the week.

```python
# Sample data - number of candies of each color
colors = ['Red', 'Blue', 'Green', 'Yellow']
counts = [5, 3, 7, 2]
```

```python
# Creating the histogram
plt.bar(colors, counts)
```

```python
# Adding labels and title
plt.xlabel('Color')
plt.ylabel('Count')
plt.title('Candy Color Histogram')
# Creating the histogram
plt.bar(colors, counts)
# Displaying the histogram
plt.show()
```
